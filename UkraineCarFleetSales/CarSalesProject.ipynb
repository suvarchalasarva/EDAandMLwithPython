### 3.3 Pre Processing

- Dealing with missing values<br/>
    - Replacing missing entries of __Drive__ with mode value for that column
    - Replacing missing values of __Engine Volume__ with median value.
 

```python
carsales_df['drive'].fillna(carsales_df['drive'].mode().values[0], inplace = True)
carsales_df['drive']
```

```python
carsales_df['drive'].isnull().sum() # Here we are replacing NaN values with mode of "drive" column and checking that '0' null values are returned
```

```python
engV_median=carsales_df['engV'].median()
carsales_df['engV'].fillna(engV_median, inplace=True)
carsales_df['engV']
```

```python
carsales_df['engV'].isnull().sum() # Here we are replacing NaN values with Median of "Engine Volume" column and returning 0 null value records
```

### Apart from Engine Volume and Drive Type, there are missing /wrong values in Mileage and Price columns too if we observe.
### Replacing zero(0) values in "Mileage" column and "Price" column with respective "Mean" values

```python
carsales_df['mileage']
```

```python
mean_val=carsales_df["mileage"].mean()
mean_val
```

```python
carsales_df["mileage"]=carsales_df["mileage"].map(lambda x: carsales_df["mileage"].mean() if x==0 else x)
carsales_df["mileage"]
```

```python
carsales_df['price']
```

If we observe, there are 0 values for few car records whereas no car can have 0 as price, so we will replace cars with 0 price with __Mean__ value

```python
mean_val_price=carsales_df["price"].mean()
mean_val_price
```

```python
carsales_df["price"]=carsales_df["price"].map(lambda x: carsales_df["price"].mean() if x==0 else x)
carsales_df["price"]
```

Here we are Segmenting __'New/Used'__ column based on __Registration__ column values, that is if car is registered;it is __Used__ car and if it is not registered yet, it would be a __New__ car

```python
 carsales_df['New/Used'] = carsales_df.apply(lambda x: 'Used' if x['registration'] =='yes' else 'New',axis=1)
carsales_df['New/Used']
```

Here we are retrieving top five cars(rows) that are __registered__ with new column __New/Used__ added to Dataframe

```python
carsales_df[carsales_df.registration=='yes'].head(5) 
```

Here we are retrieving top five cars(rows) that are __not registered__ with new column __New/Used__ added to Dataframe

```python
carsales_df[carsales_df.registration=='no'].head(5)
```

Here we are calculating count of __duplicate__ rows and __removing duplicate__ rows from our original dataframe and __updating__ the dataframe

```python
carsales_df.duplicated().sum()
```

```python
carsales_df.drop_duplicates(inplace=True)
carsales_df
```
